### 사이킷런 첫걸음 : 퍼셉트론 훈련
- 사이킷런이란?
  * 사이킷런 라이브러리는 많은 학습 알고리즘을 제공
  * 데이터 전처리, 세부조정, 모델평가를 위해 편리하게 사용할 수 있는 함수 존재





```python
from sklearn import datasets
import numpy as np
import pandas as pd
```

```python
iris = datasets.load_iris()
X = iris.data[:,[2,3]]
y = iris.target

# Convert to DataFrame for better visualization
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Display the first few rows of the dataset
iris_df.head()
print(np.unique(y))
```

```python
#모델 성능을 평가하기 위해 데이터셋을 train/test set으로 나눈다
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state =1 ,stratify =y)
```

- test set : 30%, 70%는 훈련데이터로 사용
- train_test_split 함수는 데이터를 분할하기 전 데이터를 미리 섞는다.
  - random_state를 고정하면 결과를 재현할 수 있다
- stratify : 계층화
- 훈련세트와 테스트 세트의 클래스 레이블 비율을 입력 데이터셋과 동일하게 한다

```python
#특성 스케일 조정-평균 0,  분산 1로
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
#퍼셉트론 모델로 훈련
from sklearn.linear_model import Perceptron
ppn = Perceptron(max_iter =40, eta0 = 0.01, tol =1e-3, random_state =1)
ppn.fit(X_train_std, y_train)
```

```python
y_pred = ppn.predict(X_test_std)
print('misclassified sample :%d'%(y_test != y_pred).sum())
```
